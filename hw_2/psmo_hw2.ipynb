```python
y_hat = model.predict(x).values
```

```python
xx = x[:]
xx['const'] = np.ones(x.shape[0])
```

```python
# Возьмем HC_0
from scipy.sparse import diags
from scipy.sparse import csr_matrix
u_hat_sq =  (y_hat - y.values)**2
var_u = diags(u_hat_sq)
var_b = csr_matrix(np.linalg.inv(xx.T @ xx)@ xx.T).dot(var_u).dot(csr_matrix(xx @ np.linalg.inv(xx.T @ xx)))        
```

```python
print(f'{np.diag(var_b.toarray()**0.5)}') 
```

## Задача 2: "Борьба с гетероскедастичностью" (5 баллов).

Рассмотрим линейную регрессионную модель

$$y_i = 10+2x_i+\varepsilon_i$$

$$x_i \sim N(40, 4)$$

$$\varepsilon_i \sim N(0,x_i^2)$$

**Пункт 1.** Сгенерируйте выборку размера $n=50$. Оцените модель с помощью МНК. Прокомментируйте значимость регрессоров и проверьте  модель на гетероскедастичность при помощи вашего любимого теста. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Статистика
import statsmodels.formula.api as smf  # Модели и гипотезы
import statsmodels.stats.api as sms  # Тесты
import statsmodels.api as sm
```

```python
m = 40
x = np.random.normal(40, 2, 50)
e = [np.random.normal(0, x[i]) for i in range(len(x))]
y = [10 + 2 * x[i] + e[i] for i in range(len(x))]
df = pd.DataFrame(data=np.array([x, y]).T, columns=['x', 'y'])
df.head()
```

```python
model = smf.ols('y ~ x', data=df).fit()
```

```python
model.summary()
```

**По значению p-value переменные  значимы, то же самое можно сказать по p-value F статистики о значимости регрессии, что регрессия значима.**

```python
sms.het_goldfeldquandt(model.model.endog, model.model.exog)
```

**Тест Голдфильда-Кванта не выявил гетероскедастичность, хотя стоит признать, что был относительно близок**

**Пункт 4.** Оцените стандартные ошибки коэффициентов, устойчивые к гетероскедастичности, при помощи какого-нибудь метода бутстрэпа. Сравните результаты оценки с результатами предыдущего пункта.

```python
# Взято с семинара
from statsmodels.regression.linear_model import OLS, GLS
n_reps = 100

cov_matr = np.zeros((2, 2))

for i in range(n_reps):
    # 1. Строим бутстрэп-выборку
    random_inds = np.random.randint(0, len(model.model.endog), size = len(model.model.endog))
    y_b = model.model.endog[random_inds]
    X_b = model.model.exog[random_inds, :]
    
    # 2. Оцениваем модель и обновляем ковариационную матрицу
    model_b = OLS(y_b, X_b).fit()
    coefs = np.array(model_b.params).reshape(-1, 1)
    cov_matr = cov_matr + (coefs - np.mean(coefs)) @ (coefs - np.mean(coefs)).T
```

**Пункт 5.** Оцените стандартные ошибки коэффициентов, устойчивые к гетероскедастичности, при помощи какой-нибудь из схем HC. Поменялось ли что-нибудь?

$$\hat Var(\hat \beta) = (X^TX)^{-1}X^T \hat Var(u) X(X^TX)^{-1}$$
$$\hat Var(u) = \begin{bmatrix}\hat u_1^2 & ... \\ ... & \hat u_2^2 \\  & & \ddots \end{bmatrix}$$

```python
y_hat = model.predict(df['x']).values
```

```python
np.ones(len(df['x'].values))
```

```python
xx = np.vstack([np.ones(len(df['x'].values)), df['x'].values]).T
```

```python
# Возьмем HC_0
u_hat_sq =  (y_hat - df['y'].values)**2
var_u = np.diag(u_hat_sq)
var_b = np.linalg.inv(xx.T @ xx) @ xx.T @ var_u @ xx @ np.linalg.inv(xx.T @ xx)         
```

```python
print(f'{np.diag(var_b)**0.5}') 
```

**Стандартная ошибка наклона поменялась кардинально в лучшую сторону, из-за учета гетероскедастичности в оценке.**
